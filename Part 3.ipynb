```python
Cnt_ServiceID = df[['Serviceid', 'User_id']].groupby('Serviceid').count().reset_index()
```

```python
Cnt_ServiceID.rename (columns = {'User_id':'Count'}, inplace = True)
```

```python
Cnt_ServiceID
```

```python
n_Cnt_ServiceID = Cnt_ServiceID.nlargest(10, columns = 'Count')
```

**Top 10 ID_Service có nhiều lượt truy cập nhất

```python
n_Cnt_ServiceID
```

```python
small_Cnt_ServiceID = Cnt_ServiceID.nsmallest(20, columns = 'Count')
```

**20 ID_service có lượt truy cập ít nhất

```python
small_Cnt_ServiceID
```

**Những ID_Serviced có lượt truy cập <5
